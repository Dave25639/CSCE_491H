```python
import matplotlib.pyplot as plt
import numpy as np

embeddings = ['VAE', 'ResNet, AE', 'ResNet', 'Lunit', 'AE', 'UNI', 'UNI, AE']
test_c_index = [0.5259798099, 0.5330223656, 0.5422824335, 0.6388824542, 0.6628683996, 0.7388906184, 0.7496971383]
lower_ci = [0.3530884152, 0.278042522, 0.3252636476, 0.3721296955, 0.434709596, 0.4671167889, 0.5476066323]
upper_ci = [0.6522896316, 0.7879645984, 0.7246216296, 0.865870976, 0.8789050449, 0.9230235043, 0.9016899706]

test_c_index = np.array(test_c_index)
lower_ci = np.array(lower_ci)
upper_ci = np.array(upper_ci)

fig, ax = plt.subplots(figsize=(10, 6))

ax.errorbar(test_c_index, embeddings, xerr=[test_c_index - lower_ci, upper_ci - test_c_index], fmt='o', color='black', label='95% CI')
ax.scatter(test_c_index, embeddings, color='blue', zorder=5)

for i, embedding in enumerate(embeddings):
    ax.vlines(lower_ci[i], i - 0.1, i + 0.1, color='black')  # Lower CI line
    ax.vlines(upper_ci[i], i - 0.1, i + 0.1, color='black')  # Upper CI line

ax.set_xlabel('Test c-index')
ax.set_title('Forest Plot of Test c-index with 95% Confidence Interval (Single Data Point)')

ax.grid(True)

plt.tight_layout()
plt.show()
```
